# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import pickle
pd.set_option('display.max_rows',500)
pd.set_option('display.max_columns',500)
```

# Loading the data

```python
data = pd.read_csv("train.csv")
```

```python
lval = ['PCIAT-PCIAT_01',
'PCIAT-PCIAT_02',
'PCIAT-PCIAT_03',
'PCIAT-PCIAT_04',
'PCIAT-PCIAT_05',
'PCIAT-PCIAT_06',
'PCIAT-PCIAT_08',
'PCIAT-PCIAT_09',
'PCIAT-PCIAT_10',
'PCIAT-PCIAT_11',
'PCIAT-PCIAT_13',
'PCIAT-PCIAT_14',
'PCIAT-PCIAT_15',
'PCIAT-PCIAT_16',
'PCIAT-PCIAT_17',
'PCIAT-PCIAT_18',
'PCIAT-PCIAT_19',
'PCIAT-PCIAT_20',
'Physical-Height',
'Physical-Weight','PreInt_EduHx-computerinternet_hoursday','Physical-BMI','Physical-Waist_Circumference','sii']
```

```python
data = data[lval]
```

```python
data.head()
```

# Deleting null sii values

```python
data = data.dropna(subset = ["sii"])
```

# Filling in the mean for other feature null vals

```python
for i in lval:
    data[i]=data[i].fillna(round(data[i].mean(),2))
```

```python
data.isna().sum()
```

# Setting up x and y for the model

```python
x = data[lval].drop(["sii"],axis=1)
y = data["sii"]
```

```python
x.head()
```

```python
x.shape
```

```python
y.head()
```

# Split into train and test

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

```python
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

# Scaling the data

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(xtrain)
x_test = sc.fit_transform(xtest)
```

```python
x_train.shape
```

```python
type(x_train)
```

```python
print(x_train)
```

# RFC

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 100, random_state = 42)
rfc.fit(x_train,ytrain)
```

```python
with open("sii_prediction.pkl","wb") as f:
    pickle.dump(rfc,f)
```

```python
ypred = rfc.predict(x_test)
print(ypred)
```

# Checking for accuracy

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
accuracy_score(ytest,ypred)
```

```python
print(classification_report(ytest,ypred))
```

```python
import tkinter as tk
from tkinter import messagebox
answers = [None] * len(featurequestions) #varibale that will contain all of the answers
current = 0 # track which q we are on


root = tk.Tk()
root.title("AI Powered PCIAT Test")
root.geometry("600x400")
selected_option = tk.StringVar()

questionLabel = tk.Label(root,text="", wraplength = 450)
questionLabel.pack()

option_buttons = [] #list stores the radio button widgets
entry_field = tk.Entry(root) #for storing entry widget for imput type q's

#function will display the current question
def show_question():
    #global entry_input
    global current
    #entry_input = None
    selected_option.set("")
    questionLabel.config(text=f"Q{current+1}:  {featurequestions[current]["question"]}") #displaying the question


    for button in option_buttons:
        button.destroy()

    option_buttons.clear()

    



    
    entry_field.pack_forget() #this hides the input field if it was there preiousloy
    #displaying the options
    if featurequestions[current]["type"]=="mcq":
        for option in featurequestions[current]["options"]:
            btn = tk.Radiobutton(root,text=option,variable=selected_option, value=option)
            btn.pack(anchor="w")
            option_buttons.append(btn)
        if answers[current] != None:
            selected_option.set(str(answers[current])) #var where answer given by user is stored
            
    else:
        entry_field.pack() #displaying entry box
        entry_field.delete(0,tk.END) #deleting anything written in the entry box
        if answers[current]:
            entry_field.insert(0,str(answers[current])) #simply inserting the answer into the entry field
        #entry_input = entry_field

def next_question():
    global current
    if featurequestions[current]["type"]=="mcq":
        option = selected_option.get() #insering the answer into the option variable
        if option == "":
            messagebox.showwarning("WARNING: Please select an option before proceeding")
            return
            
        answers[current] = int(option) #appending the answer into the answers list
    else:
        option = entry_field.get().strip()
        if option == "":
            messagebox.showwarning("WARNING: Please select an option before proceeding")
            return
            
        answers[current] = float(option)
        
    
    if current < len(featurequestions)-1:
        current = current + 1
        show_question()

    else:
        show_summary()

def prev_question():
    global current
    if current > 0: #making sure its not the first q and then moving to the prev q
        current = current -1 
        show_question()

def show_summary():
    questionLabel.config(text="Quiz Completed")


    entry_field.pack_forget() #this hides the input field if it was there preiousloy
    summary = tk.Text(root, height = 20, width = 60)
    summary.pack()
    #loading the model
    with open("sii_prediction.pkl","rb") as f:
        rfc = pickle.load(f)

    #using model here to give sii score
    text = "The sii score is: " + str(rfc.predict([answers]))+ "\n"
    summary.insert(tk.END,text)


    for index,i in enumerate(answers):
        text = "The answer you provided for question: " + str(index) + " is: " + str(i)+ "\n"
        summary.insert(tk.END,text)
    print(answers)
frame = tk.Frame(root)
frame.pack()


prev_btn = tk.Button(frame,text="Back",command=prev_question)
prev_btn.pack()


next_btn = tk.Button(frame,text="Next", command=next_question)
next_btn.pack()

show_question()

root.mainloop()
```
